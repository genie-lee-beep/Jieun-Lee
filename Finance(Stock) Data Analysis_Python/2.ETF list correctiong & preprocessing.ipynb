### HTTP request using 'requests'
* [Requests: HTTP for Humans™ — Requests documentation](https://requests.readthedocs.io/en/master/)

```python
url= 'https://finance.naver.com/api/sise/etfItemList.nhn'
```

```python
# requests 라이브러리를 통해 url을 받아옵니다.
# request는 키와 값이 있다.
response = requests.get(url)
response
```

### JSON

```python

etf_json = response.json()
```

```python
etfItemList = etf_json['result']['etfItemList']
etfItemList[0]
```

## Converting to DataFrame
### Converting JSON data to a DataFrame in pandas"

## data preprocessing

```python
eftcode = """전체
국내 시장지수
국내 업종/테마
국내 파생
해외 주식
원자재
채권
기타"""
eftcode
```

```python
etf_tab_name = eftcode. split()
etf_tab_name
```

```python
def find_etf_tab_name(no):
    return etf_tab_name[no]
```

```python
find_etf_tab_name(2)
```

```python
df['etfTabName'] = df['etfTabCode'].map(find_etf_tab_name)
```

### 파생변수 만들기
* 브랜드, 인버스, 레버리지, 환헤지H 변수 만들기
* [Working with text data — pandas documentation](https://pandas.pydata.org/docs/user_guide/text.html)

```python
df["인버스"]= df['itemname'].str.contains("인버스")
```

```python
df["레버리지"] = df['itemname'].str.contains("레버리지")
```

```python
# (H)의 의미는 환율변동 위험을 막기 위해 Hedge를 한다는 뜻입니다.
df["환헤지H"] = df['itemname'].str.contains("H\)")
                
```

```python
df.loc[df["itemname"].str.contains("H"),['itemname',"환헤지H"]]
```
