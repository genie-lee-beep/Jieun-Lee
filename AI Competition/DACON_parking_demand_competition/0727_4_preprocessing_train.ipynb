```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder

cols = ['총세대수_lb_A1','총세대수_lb_B1','총세대수_lb_C1', '총세대수_lb_D1']
for col in cols :

    lb_encoder =LabelEncoder()
    lbs = lb_encoder.fit_transform(train[col])
    
    train['총세대수_lb_A1','총세대수_lb_B1','총세대수_lb_C1', '총세대수_lb_D1'] = lbs # all_df에 col+'_lb' 열 추가
    lbs = lbs.reshape(-1,1) 
    col_names = lb_encoder.classes_
    col_dict = {idx : name for idx, name in enumerate(col_names)} # all_df 칼럼명 설정하기 위해 dict 형태로 저장
    print(col_dict) 
    
    oh_encoder = OneHotEncoder()
    ohs = oh_encoder.fit_transform(lbs).toarray() 
    
    train = pd.concat([train, pd.DataFrame(ohs)], axis=1)
    train.rename(columns=col_dict, inplace=True)
    
print(train.columns)
train
```
