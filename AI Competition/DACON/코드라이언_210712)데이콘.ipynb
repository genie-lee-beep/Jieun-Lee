```python
# 변수를 하나 >>총세대수 
sel = ["총세대수"]
X_train = train[sel]
X_test = test[sel]
y_train = train['등록차량수']
```

```python
from sklearn.linear_model import LinearRegression   # y= a*총세대수 +b
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
pred = model.predict(X_test)
pred
```

```python
test['등록차량수'] = pred
test['코드별차량수평균'] = test.groupby('단지코드')['등록차량수'].transform(np.mean)
test_new = test.drop_duplicates(['단지코드'], keep='first').reset_index()
test_new
```

```python
#code와 num으로 제출을 하라고 써있다

sub_df = test_new[ ['단지코드', '코드별차량수평균']]
sub_df.columns = ['code', 'num']
sub_df
```
