```python
import pandas as pd
```

```python
subtitle_data = pd.read_csv('youtube_subtitle.csv')
```

```python
df = pd.DataFrame(subtitle_data)
df
```

```python
df_plus = df.assign(g=df.x % 2)
df_plus
```

```python
df_subtitle = df.assign(g=df.x % 2).query('g==1')
df_subtitle 
```

```python
subtitle= df_subtitle [['y']]
subtitle
```

```python
subtitle_outcome=subtitle.set_index('y')
subtitle_outcome
```

```python
subtitle_out_df = pd.DataFrame(subtitle_outcome)
subtitle_out_df
```

```python
subtitle_out_df.to_csv('youtube_subtitle_out.csv', index=True)

# subtitle_out_df.to_csv('filename.csv')
```
