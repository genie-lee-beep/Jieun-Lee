### 총세대수 자료형을 numeric으로 변경

원해 object 자료형이었던 '총세대수'자료형을 숫자형으로 바꿔준다.

* 참고: https://sikaleo.tistory.com/39

```python
test['총세대수'] = pd.to_numeric(test['총세대수'])
print('test')
```

```python
test.dtypes
```

```python
test['총세대수'].unique()
```

```python
test['총세대수_lb']=0
```

```python

test.loc[test['총세대수'] <=149 ,'총세대수_lb'] = 'A1'
test.loc[(test['총세대수'] > 149) & (test['총세대수'] <= 299),'총세대수_lb' ] = 'B1'
test.loc[(test['총세대수'] > 299) & (test['총세대수'] <= 499), '총세대수_lb' ] = 'C1'
test.loc[(test['총세대수'] > 499) & (test['총세대수'] <= 999), '총세대수_lb' ] = 'D1'
test.loc[(test['총세대수'] >= 1000), '총세대수_lb' ] = 'E1'
```

```python
all_test= pd.concat([test, col_dummies], axis = 1)
all_test
```
