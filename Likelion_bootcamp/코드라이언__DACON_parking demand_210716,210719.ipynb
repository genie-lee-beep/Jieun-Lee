```python
print(train.ìê²©ìœ í˜•.unique())
print(test.ìê²©ìœ í˜•.unique())

mapping = { 'A':1, 'B':2, 'C':3, 'D':4, 'E':5,
            'F':6, 'G':7, 'H':8, 'I':9, 'J':10,
            'K':11, 'L':12, 'M':13, 'N':14, 'O':15  }

train['ìê²©ìœ í˜•'] =train['ìê²©ìœ í˜•'].map(mapping).astype(int)
test['ìê²©ìœ í˜•'] =test['ìê²©ìœ í˜•'].map(mapping).astype(int)

train.head(3)
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
sel = ['ì´ì„¸ëŒ€ìˆ˜', 'ì „ìš©ë©´ì ', 'ì „ìš©ë©´ì ë³„ì„¸ëŒ€ìˆ˜',
       'ê³µê°€ìˆ˜',  'ë‹¨ì§€ë‚´ì£¼ì°¨ë©´ìˆ˜', 'ìê²©ìœ í˜•']
X = train[sel]
y = train['ë“±ë¡ì°¨ëŸ‰ìˆ˜']
```

```python
# ë°ì´í„° ë‚˜ëˆ„ê¸°
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,
                                                   random_state=0)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
pred = model.predict(X_test)
```

```python
print("í•™ìŠµ(score) :", model.score(X_train, y_train) ) # ê²°ì •ê³„ìˆ˜
print("í…ŒìŠ¤íŠ¸(score) :", model.score(X_test, y_test) ) # ê²°ì •ê³„ìˆ˜
# mae
np.mean( np.abs(y_test - pred)  )
# mse
mse_val = np.mean( (y_test - pred)**2  )
mse_val
# rmse
np.sqrt(  mse_val )
```
