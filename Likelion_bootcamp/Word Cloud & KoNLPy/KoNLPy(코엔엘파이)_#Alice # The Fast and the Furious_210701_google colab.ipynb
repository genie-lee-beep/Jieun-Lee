## 분노의 질주- 한국어 리뷰

```python
doc_ko = open("15_TheExtreme_utf8.txt").read()
doc_ko[1:1000]

# 데이터가 너무 많으면 read.line으로?? 
```

```python
# OKT 클래스를 이용한 명사확인 Okt (11초 : 4286), Kkma (37초 : 5572)
from konlpy.tag import Okt ### Okt를 활용하여 명사를 가져옴
t = Okt()
#from konlpy.tag import Kkma ### Okt
#t = Kkma()
doc_nouns = t.nouns(doc_ko)
print( len( doc_nouns ) )
doc_nouns[0:30]   #리스트니까 슬라이싱이 가능하겠다 예상할 수 있음
```

### 중복 단어는 제외하는 방법


```python
# nltk.Text()를 이용하여 nltk가 가지는 많은 기능을 사용 가능해짐.
ko = nltk.Text(doc_nouns, name="분노의 질주")
print(type(ko))
print(len(ko.tokens))   #명사 하나하나가 token
```

```python
### 단어들의 사용 횟수 확인 - 빈도 분석
type( ko.vocab() ), ko.vocab().most_common(50)
```

```python
### 중복된 단어를 제거한 개수를 확인     질문: 중복되었다는게 어디에 나옴?
print(len(set(ko.tokens)))
```

```python
plt.figure(figsize=(12, 6))
ko.plot(50)
plt.show()
```

```python
### 한글에서는 따로 불용어 사전 라이브러리 없어, 따로 만들거나 또는 파일로 부터 불러올 수 있다.
stop_words = ['분노', '영화', '액션', '시리즈', '더',
'그', '이', '것', '또', '좀',
'돈', '것', '다음', '질주', '그냥',
'분노의질주', '말', '뭐', '애', '나', '듯', '편', '볼', '점', '중', '로']
new_ko = [ ]   #여기서 부터가 한글의 경우에 추가되는 부분
for one_word in ko:
  if one_word not in stop_words:  # 만약 one_word가 stop_word에 없으면 new_ko에 더하겠다. 
   new_ko.append(one_word)



#-----------
#[영어는 이미 불용어 단어 라이브러리를 가져오는 거]
### 불용어 단어 추가
#stopwords = set(STOPWORDS)
#stopwords.add("said")
#print( len(stopwords) ) # 190 -> 191로 변경
#stopwords
```

```python
### nltk Text 객체 만들기
new_ko = nltk.Text(new_ko, name="분노의 질주")
plt.figure(figsize=(12,6))
new_ko.plot(50)
```

### 실습_나만의 불용어 사전 만들기
* new_ko_2 = [ ] 

```python
f = open("stopwords_extreme(movie).txt")   #메모장에 나만의 불용어 단어들을 저장함
lines = f.readlines()
stop_words = []
for one in lines:
  stop_words.append(one.strip())
new_ko_2 = [ ]
for one_word in new_ko:
  if one_word not in stop_words:
    new_ko_2.append(one_word)
```
