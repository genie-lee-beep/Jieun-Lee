## ë°ì´í„° ëª¨ë¸ë§

```python
from sklearn.linear_model import LinearRegression
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
# ë‹¤ìˆ˜ì„ í˜•íšŒê·€
sel = ['ì´ì„¸ëŒ€ìˆ˜', 'ì „ìš©ë©´ì ', 'ì „ìš©ë©´ì ë³„ì„¸ëŒ€ìˆ˜',
       'ê³µê°€ìˆ˜',  'ë‹¨ì§€ë‚´ì£¼ì°¨ë©´ìˆ˜', 'ìê²©ìœ í˜•']
X = train[sel]
y = train['ë“±ë¡ì°¨ëŸ‰ìˆ˜']
```

```python
# ë°ì´í„° ë‚˜ëˆ„ê¸°
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,
                                                   random_state=0)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
pred = model.predict(X_test)
```

```python
print("í•™ìŠµ(score) :", model.score(X_train, y_train) ) # ê²°ì •ê³„ìˆ˜
print("í…ŒìŠ¤íŠ¸(score) :", model.score(X_test, y_test) ) # ê²°ì •ê³„ìˆ˜

# mae
mae = np.mean( np.abs(y_test - pred)  )
print('mae:',mae)

# mse
mse_val = np.mean( (y_test - pred)**2  )
print('mse:', mse_val)

# rmse
np.sqrt(  mse_val )
```
