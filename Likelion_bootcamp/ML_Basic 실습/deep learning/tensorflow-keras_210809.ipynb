```python
train = pd.read_csv("/kaggle/input/bike-sharing-demand/train.csv", parse_dates=['datetime'])
test = pd.read_csv("/kaggle/input/bike-sharing-demand/test.csv", parse_dates=['datetime'])
sub = pd.read_csv("/kaggle/input/bike-sharing-demand/sampleSubmission.csv")
```

```python
input_col = ['temp','humidity']
labe_col = ['count']

X = train[input_col]
y = train[labe_col]
X_val = test[input_col]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, random_state=0)
print(train.shape)
print(test.shape)
```

```python
from keras.models import Sequential
from keras.layers import Dense 
```

```python
model = Sequential()
model.add(Dense(1, input_dim =2, activation='sigmoid'))
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(30, input_dim=2, activation='relu'))
model.add(Dense(15, activation='relu') )
model.add(Dense(15, activation='relu') )
model.add(Dense(1))
```

```python
model.summary()
```

```python
model.compile(loss = 'mean_squared_error', optimizer='rmsprop')
model.fit(X_train, y_train, epochs=20, batch_size=10)

# 좋아진 상태부터 다시 돌린다는 점이 sklearn이랑 다른 점
```

```python
### 평가 확인
model.evaluate(X_test, y_test)
```

```python
pred = model.predict(X_val)
```

```python
sub = pd.read_csv("/kaggle/input/bike-sharing-demand/sampleSubmission.csv")
sub['count'] = pred

sub.loc[sub['count']<0, 'count'] = 0
```
