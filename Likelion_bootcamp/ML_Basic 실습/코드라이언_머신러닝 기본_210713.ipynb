```python
tips = sns.load_dataset("tips")
tips
```

### 3-2-1 데이터 나누기

```python
from sklearn.model_selection import train_test_split

sel = ["total_bill", "tip"]
X = tips[sel]
y = tips['size']

X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=0) #데이터를 뽑을 때 마다 동일한 데이터를 뽑기 위해. 뽑을 때 마다 다른 데이터를 뽀으면 평가할 때 안 좋으니까      
```

3-2-2 knn 모델 만들기

```python
from sklearn.neighbors import KNeighborsClassifier  #사이즈니까 classifier
knn = KNeighborsClassifier(n_neighbors=1)  # t(  )은 ap누르면 완성
```

```python
knn.fit(X_train, y_train)
```

```python
pred = knn.predict(X_test)
pred
```

### 3-2-3 평가해 보기

```python
np.mean(pred ==y_test)
```

### 좋은 k를 찾기

```python
good_k = 999  #좋은 k 찾기
accuracy = 0  #정확도
for i in range(1,50,1):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    pred = knn.predict(X_test)
    
    
    if accuracy <= np.mean(pred ==y_test):
        good_k = i                             #k값 업그레이드
        accuracy = np.mean(pred ==y_test)      #정확도 업그레이드
        print(np.mean(pred ==y_test))
```
