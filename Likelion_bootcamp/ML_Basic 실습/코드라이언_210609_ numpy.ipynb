### 추가(한그래프위에 cos,sin, tan) 그래프 그리기

```python
x1 = np.linspace(0, 14, 50)
y2 = np.cos(x1)

x2 = np.linspace(0, 14, 50)
y2 = np.tan(x2)

x3 = np.linspace(0, 14, 50)
y3 = np.sin(x3)

plt.plot(x1,y2, 'g^', x2,y2,'o', x3,y3,'^') #plt.plot(x,y,'^') 삼각형 점, plt.plot(x,y,'o') 점선
plt.show()
```

```python
x1 = np.linspace(0, 14, 50)
y2 = np.cos(x1)

x2 = np.linspace(0, 14, 50)
y2 = np.tan(x2)

x3 = np.linspace(0, 14, 50)
y3 = np.sin(x3)

plt.plot(x1,y2, 'g^', x2,y2,'o', x3,y3,'^') 
plt.title('my graph')
plt.xlabel('a','b','c')
plt.ylabel('1','2','3')
plt.show()
```
