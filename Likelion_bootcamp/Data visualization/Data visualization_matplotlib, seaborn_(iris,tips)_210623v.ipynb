```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")  # 인터넷이 켜져 있어야 함.
tips
```

```python
tips.columns
```

```python
# 요일별 팁 받은 횟수 출력
sns.countplot(x='day',data=tips)
```

```python
iris_size = iris.pivot_table(index='sex'
            ,columns='day' ,aggfunc='size') 
# aggfunc='size' 각 데이터의 건수에 대해 출력
# 그냥 pivot table은 평균, 분산 등의 결과가 출력됨
iris_size
```

```python
### 요일별 식사 금액은 얼마나 될까?
plt.figure(figsize=(8,6))
sns.barplot(x="day", y="total_bill", data=tips)   #  sns. 어떤 plot을 할건지(x와 y 컬럼을 써주고, data = 어떤 데이터를 쓸건지)
plt.show()
```

```python
tips.columns
```

```python
tips.info()
```

```python
print(sns.__version__)
sns.barplot(x='sex', y='tip', data=tips) # barplot은 x축 (범주형 )y축이 숫자 (범주형 그래프)   # Y축이 문자열이면 countplot? 
```

```python
### 요일별로 팁은 얼마나 될까?

plt.figure(figsize=(8,6))
sns.barplot(x="day", y="tip", data=tips)
plt.show()
```

```python
sns.boxplot(x = 'day', y = 'total_bill',data=tips,
            hue='smoker', palette='Set3')
# hue = 'smoker'
# hue 옵션을 사용해서 카테고리별 비교가 가능

# palette='Set3'
# palette 색상 옵션
```

```python
### 여성 남성 흡연 
tips.sex = tips.sex.map({'female':0, 'male':1})
tips.smoker = tips.smoker.map({'No':0, 'Yes':1})
tips.day = tips.day.map({'Thur':0, 'Fri':1, 'Sat':2, 'Sun':3})
```

```python
tips.info()
```

```python
## 토요일날 담배피는 남자가 와서 팁을 낼 확률
```

```python
sns.lmplot(x="total_bill", y="tip", data=tips)
```

## 2. iris

```python
import seaborn as sns
import matplotlib.pyplot as plt
iris = sns.load_dataset("iris")
iris.head()
```

```python
iris.info()
```

```python
iris.columns
```

```python
iris.tail(10)
```

```python
iris.describe()
```

```python
sns.barplot(x='species', y='sepal_length',data = iris)
```

```python
sns.barplot(x="species", y="sepal_length", hue="petal_length", data=iris)  # hue는 시간이 너무 걸려서, 권하지 않음
```

```python
sns.pairplot(iris)
plt.show()
```

```python
# hue 옵션을 이용해서 종별 특성을 색상으로도 알아보기
sns.pairplot(iris, hue='species', palette='husl')
plt.show()
```

```python
# 몇 개의 칼럼만 확인하기
sns.pairplot(iris, height=3, vars=['sepal_length','sepal_width'])
plt.show()
```

### 4개 feature 시각화 해보기

```python
##이예준님 코드
fig = plt.figure(figsize=(10,6))
fig.patch.set_facecolor('xkcd:white')
axis1 = fig.add_subplot(2, 2, 1)
axis1 = sns.barplot(x='species', y='sepal_length', data=iris)
axis1.plot()
axis2 = fig.add_subplot(2, 2, 2)
axis2 = sns.barplot(x='species', y='sepal_width', data=iris)
axis2.plot()
axis3 = fig.add_subplot(2, 2, 3)
axis3 = sns.barplot(x='species', y='petal_length', data=iris)
axis3.plot()
axis4 = fig.add_subplot(2, 2, 4)
axis4 = sns.barplot(x='species', y='petal_width', data=iris)
axis4.plot()
```
