```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
x =np.linspace(0,14,100) # 0부터 14까지 동일한 간격의 값 100개 만들기
x
```

```python
x =np.linspace(0,14,100) # 0부터 14까지 동일한 간격의 값 100개 만들기

print(type(x),len(x))
y1 =np.sin(x)  # x의 값을 이용하여 y값 (sin) 구하기
y2 = 2*np.sin(x+0.5)
y3 = 4*np.sin(x+1.0)
plt.figure(figsize=(10,6))   # 그림의 크기
plt.plot(x, y1)
```

```python
plt.plot(x, y1, x,y2, x, y3)   # 3개의 sin 그래프
```

```python
# 위의 그래프에, 범례달기
plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='2*sin(x)')
plt.plot(x, y3, label='4*sin(x)')

plt.legend()
```

```python
### 요일별로 팁은 얼마나 될까?

plt.figure(figsize=(8,6))
sns.barplot(x="day", y="tip", data=tips)
plt.show()
```

```python
### 여성 남성 흡연 
tips.sex = tips.sex.map({'female':0, 'male':1})
tips.smoker = tips.smoker.map({'No':0, 'Yes':1})
tips.day = tips.day.map({'Thur':0, 'Fri':1, 'Sat':2, 'Sun':3})
```

```python
##이예준님 코드
fig = plt.figure(figsize=(10,6))
fig.patch.set_facecolor('xkcd:white')
axis1 = fig.add_subplot(2, 2, 1)
axis1 = sns.barplot(x='species', y='sepal_length', data=iris)
axis1.plot()
axis2 = fig.add_subplot(2, 2, 2)
axis2 = sns.barplot(x='species', y='sepal_width', data=iris)
axis2.plot()
axis3 = fig.add_subplot(2, 2, 3)
axis3 = sns.barplot(x='species', y='petal_length', data=iris)
axis3.plot()
axis4 = fig.add_subplot(2, 2, 4)
axis4 = sns.barplot(x='species', y='petal_length', data=iris)
axis4.plot()
```
