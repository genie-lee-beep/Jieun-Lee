```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
# species 컬럼을 선택합니다.
# 컬럼의 값과 조건을 비교합니다.
# 그 결과를 새로운 변수에 할당합니다.
is_setosa= df['species'] == 'setosa'


# 조건를 충족하는 데이터를 필터링하여 새로운 변수에 저장합니다.
setosa = df[is_setosa]

# 결과를 출력합니다.
setosa
```

### 두가지 조건을 만족하는 데이터 필터링하기

1) petal_length 컬럼의 값이 1.5 보다 작으면서 2) sepal_length 컬럼의 값이 5 보다 이상인 데이터를 필터링하겠습니다. 그 결과를 새로운 변수에 할당한 뒤 데이터를 출력하겠습니다.  

```python
# petal_length 컬럼을 선택합니다.
# 컬럼의 값과 조건을 비교합니다.
# 그 결과를 새로운 변수에 할당합니다.
petal_length_smaller = df['petal_length'] <= 1.5

# sepal_length 컬럼을 선택합니다.
# 컬럼의 값과 조건을 비교합니다.
# 그 결과를 새로운 변수에 할당합니다.
sepal_length_greather_5 = df['sepal_length'] >= 5

# 두가지 조건를 동시에 충족하는 데이터를 필터링하여 새로운 변수에 저장합니다. (AND)
new_df = df[petal_length_smaller & sepal_length_greather_5]

# 결과를 출력합니다.
new_df
```

### 제외 조건을 만족하는 데이터 필터링하기

* species 컬럼의 값이 setosa가 아닌 데이터를 필터링하겠습니다.   
그 결과를 새로운 변수에 할당한 뒤 상위 5개의 행을 출력하겠습니다.

```python
# species 컬럼을 선택합니다.
# 컬럼의 값과 조건을 비교합니다.
# 그 결과를 새로운 변수에 할당합니다.
not_setosa = df['species'] != 'setosa'

# 조건를 충족하는 데이터를 필터링하여 새로운 변수에 저장합니다.
subset_df = df[not_setosa]

# 결과를 출력합니다.
print(subset_df.shape)
print()

subset_df.head()
```
