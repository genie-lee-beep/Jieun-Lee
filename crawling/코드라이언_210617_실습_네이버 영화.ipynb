```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
url = 'https://movie.naver.com/movie/running/current.nhn'

page = urlopen(url)
soup = BeautifulSoup(page, "lxml")
```

영화제목만 가져오기 

```python
title_all = soup.find_all('dt', class_="tit")
title_all[2].find("a").text  #find_all은 리스트를 검색하는 거니까, []를 써줌.
```

```python
movie_title = []
for one in title_all:
  one_movie = one.find("a").text
  movie_title.append(one_movie)
movie_title
```

## 4-2 (실습) 네이버 영화 평점 가져오기



```python
rate = soup.find_all("div", class_="star_t1")
rate[3].find("span", class_="num").text
```

```python
rate_list = []
for one in rate:   # rate 리스트 안에서 정보를 하나씩 뺴와서 one에 집어 넣는다.
  one_rate = one.find("span").text
  rate_list.append(one_rate)
rate_list
```

## 4-3 (실습) 네이버 영화 참여 인원 가져오기

```python
 people = soup.find_all("span", class_="num2")
 people[1].find("em").text
# 양효진님 >> participants = i.find("div",class_="star_t1").find("span",class_="num2").find("em").text
```

```python
people_list = []
for one in people:
  one_people = one.find("em").text
  people_list.append(one_people)
people_list
```

## 4-4 (실습) 네이버 영화 예매율 가져오기

```python
booking_rate = soup.find_all("div", class_="star_t1 b_star")
booking_rate[0].find("span", class_="num").text
```

```python
booking_rate_list = []
for one in booking_rate: # booking_rate 리스트 안에서 정보를 하나씩 뺴와서 one에 집어 넣는다.
  one_booking = one.find("span").text
  booking_rate_list.append(one_booking)
booking_rate_list
```
